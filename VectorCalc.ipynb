```python
# Assign vector directions
a = 2
b = -1

qa = 1
qb = -1

def wire_B_field(location, I, r):
    """
    Get the Mangetic Field given:
        I : the current (a scalar)
        location : the location of the wire.
        r : XY unit vector field (3d numpy array)
    """

    rb  = r - location[:,None,None]
    Bx  =  I * rb[1]/np.linalg.norm(rb, axis=0)**2 #( rb[0]**2 + rb[1]**2 ) #np.linalg.norm(r, axis=0)
    By  = -I * rb[0]/np.linalg.norm(rb, axis=0)**2

    return np.asarray([Bx, By])





Bx  =  qa *(Y+1)/( (X+a)**2 + (Y+1)**2 ) - qa *(Y-1)/( (X-a)**2 + (Y-1)**2 ) #+ qb * (X+b)/( (X+b)**2 + Y**2 )
By  = -qa *(X+a)/( (X+a)**2 + (Y+1)**2 ) + qa *(X-a)/( (X-a)**2 + (Y-1)**2 )  #+ qb * (Y)/( (X+b)**2 + Y**2 )

BBx  =  qa *(Y+1)/( (X+a)**2 + (Y+1)**2 ) - qa *(Y-1)/( (X-a)**2 + (Y-1)**2 ) #+ qb * (X+b)/( (X+b)**2 + Y**2 )
BBy  = -qa *(X+a)/( (X+a)**2 + (Y+1)**2 ) + qa *(X-a)/( (X-a)**2 + (Y-1)**2 ) 

loc1 = np.array([10,15])
lok1 = np.array([10,35])
loc2 = np.array([0,15])
lok2 = np.array([0,35])
loc3 = np.array([-10,15])
lok3 = np.array([-10,35])
Bxy = wire_B_field(loc1, 1., XY) + wire_B_field(lok1, -1., XY) + wire_B_field(loc2, 1., XY) + wire_B_field(lok2, -1., XY) + wire_B_field(loc3, 1., XY) + wire_B_field(lok3, -1., XY)

### make it such that loc can be a 2d array and so can I!!

curve = np.array([[0,0],[-1, -1],[-1, -2],[-3,-3],[-4,-2],[-3,0]]).T
# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(XY[0],XY[1],Bx,By, density=1.4, linewidth=None, color='#A23BEC')
plt.plot(a,1,'-or')
plt.plot(-a,-1,'-ob')
plt.plot(curve[0],curve[1],'-og')
plt.title('Electromagnetic Field')

# Show plot with grid
plt.grid()
plt.show()

# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(XY[0],XY[1],Bxy[0],Bxy[1], density=1.4, linewidth=None, color='#A23BEC')
plt.title('Electromagnetic Field')

# Show plot with grid
plt.grid()
plt.show()
```

```python
import sys
from matplotlib.patches import Circle


### https://scipython.com/blog/visualizing-the-earths-magnetic-field/
# Mean magnitude of the Earth's magnetic field at the equator in T
B0 = 3.12e-5
# Radius of Earth, Mm (10^6 m: mega-metres!)
RE = 6.370
# Deviation of magnetic pole from axis
α = 0. #np.radians(9.6)

def B(r, theta):
    """Return the magnetic field vector at (r, theta)."""
    fac = B0 / r**3 ## RE
    return -2 * fac * np.cos(theta + alpha), -fac * np.sin(theta + alpha)

# Grid of x, y points on a Cartesian grid
nx, ny = 64, 64
XMAX, YMAX = 40, 40
x = np.linspace(-XMAX, XMAX, nx)
y = np.linspace(-YMAX, YMAX, ny)
X, Y = np.meshgrid(x, y)

def dipole_B(location, B0, X,Y, α=0):

    X = X-location[0]
    Y = Y-location[1]
    r, θ = np.hypot(X, Y), np.arctan2(Y, X)
    # Magnetic field vector
    Br = -B0 * 2 * np.cos(θ + α) / r**3
    Bθ = -B0 *     np.sin(θ + α) / r**3
    # Transform to Cartesian coordinates: NB make North point up, not to the right.
    c, s = np.cos(np.pi/2 + θ), np.sin(np.pi/2 + θ)

    return np.asarray([Br * c - Bθ * s, Br * s + Bθ * c]) 

X = X-10
Y = Y-20
##### this is in the def!!!! 
r, θ = np.hypot(X, Y), np.arctan2(Y, X)
# Magnetic field vector, B = (Ex, Ey), as separate components
Br = -B0 * 2 * np.cos(θ + α) / r**3
Bθ = -B0 *     np.sin(θ + α) / r**3
# Transform to Cartesian coordinates: NB make North point up, not to the right.
c, s = np.cos(np.pi/2 + θ), np.sin(np.pi/2 + θ)
Bxy = np.asarray([Br * c - Bθ * s, Br * s + Bθ * c]) 




fig, ax = plt.subplots()

# Plot the streamlines with an appropriate colormap and arrow style
color = 2 * np.log(np.hypot(Bx, By))
ax.streamplot(x, y, Bxy[0], Bxy[1], color=color, linewidth=1, cmap=plt.cm.inferno,
              density=1.4, arrowstyle='->', arrowsize=1.5)

# Add a filled circle for the Earth; make sure it's on top of the streamlines.

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_xlim(-XMAX, XMAX)
ax.set_ylim(-YMAX, YMAX)
ax.set_title("Magnetic Dipole")
ax.set_aspect("equal")
plt.show()


loc = np.array([-10,-20])
BBxy = dipole_B(loc, 1.e-6, X,Y, α=0)

# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,BBxy[0],BBxy[1], density=1.4, linewidth=None, color="#A23BEC")
plt.title("Magnetic Dipole Field")

# Show plot with grid
plt.grid()
plt.show()
```
